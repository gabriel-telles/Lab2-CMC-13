### Algoritmo K-means

Gerando a tupla de valores a serem agrupados num cluster

```python
t = list(zip(df['huml'],df['humw']))
plt.scatter(df['huml'], df['tibl'])
```

```python
from sklearn.cluster import KMeans

sse = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(t)
    sse.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), sse)
plt.xticks(range(1,11))
plt.xlabel('Número de clusters')
plt.ylabel('SSE')
plt.show()
```
